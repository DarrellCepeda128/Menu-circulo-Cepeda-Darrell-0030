```python
import math

# Elaborada por Cepeda Darrell
# Función para calcular el área del círculo
def areacirculo():
    print("\n Cálculo del Área del Círculo")
    print("Fórmula: A = π * r²")
    # Se pide el radio aquí dentro como se solicitó
    radio = float(input("Por favor, ingresa el radio (r): "))
    area = math.pi * (radio ** 2)
    print(f"✅ El área del círculo es: {area:.2f}")
    input("\nPresiona Enter para continuar...")

# Función para calcular el perímetro (circunferencia) del círculo
def perimetrocirculo():
    print("\n🔵 Cálculo del Perímetro del Círculo")
    print("Fórmula: P = 2 * π * r")
    radio = float(input("Por favor, ingresa el radio (r): "))
    perimetro = 2 * math.pi * radio
    print(f"✅ El perímetro del círculo es: {perimetro:.2f}")
    input("\nPresiona Enter para continuar...")

# Función para calcular el volumen de la esfera
def volumenesfera():
    print("\n🔵 Cálculo del Volumen de la Esfera")
    print("Fórmula: V = (4/3) * π * r³")
    radio = float(input("Por favor, ingresa el radio (r): "))
    volumen = (4/3) * math.pi * (radio ** 3)
    print(f"✅ El volumen de la esfera es: {volumen:.2f}")
    input("\nPresiona Enter para continuar...")   Q

# Función principal del menú
def menu():
    print("\n" + "="*40)
    print("       📐 GEOMETRÍA INTERACTIVA 📐       ")
    print("="*40)
    print("¡Hola! Selecciona una opción para calcular:")
    print("1️⃣  Calcular Área del Círculo (A = πr²)")
    print("2️⃣  Calcular Perímetro del Círculo (P = 2πr)")
    print("3️⃣  Calcular Volumen de la Esfera (V = 4/3πr³)")
    print("4️⃣  Salir")
    print("="*40)

    opcion = int(input("👉 Escribe el número de tu opción: "))

    # Lógica para decidir qué hacer
    if opcion == 1:
        areacirculo()
    elif opcion == 2:
        perimetrocirculo()
    elif opcion == 3:
        volumenesfera()
    elif opcion == 4:
        print("\n👋 ¡Gracias por usar el programa! Hasta luego. 👋")
        return True # Retorna verdadero para indicar que se debe salir
    else:
        print("\n❌ Opción no válida, intenta de nuevo.")
        input("\nPresiona Enter para continuar...")
        return False

# Programa Principal
def main():
    salir = False
    # Bucle while para que el programa siga corriendo hasta que el usuario quiera salir
    while not salir:
        salir = menu()

if __name__ == "__main__":
    main()
```
